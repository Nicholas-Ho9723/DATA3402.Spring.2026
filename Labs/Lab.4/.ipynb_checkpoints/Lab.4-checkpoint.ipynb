1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.current_value = 0

    def inc(self):
        if self.current_value < self.max_value:
            self.current_value += 1
        else:
            print(f"Error: Counter has already reached its maximum value of {self.max_value}.")

    def reset(self):
        self.current_value = 0
        
    def get_value(self):
        return self.current_value
```

```python
def test_counter():
    
    my_counter = Counter(max_value=3)
    print(f"Initial value: {my_counter.get_value()}")
    
    my_counter.increment()
    print(f"1 increment: {my_counter.get_value()}")

    my_counter.increment()
    print(f"2 increment: {my_counter.get_value()}")
    
    my_counter.increment()
    print(f"3 increment: {my_counter.get_value()}")
    
    my_counter.increment() 
    
    my_counter.reset()
    print(f" Reset Value: {my_counter.get_value()}")
    
if __name__ == "__main__":
    test_counter()
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class counter:
    def __init__(self, max_value):
        self.__max_value = max_value
        self.__current_value = 0

    def increment(self):
        if not self.is_at_max():
            self.__current_value += 1
        else:
            print(f"Error: Counter has already reached its maximum value of {self.__max_value}.")

    def reset(self):
        self.__current_value = 0
        
    def get_value(self):
        return self.__current_value
        
    def get_max_value(self):
        return self.__max_value
        
    def is_at_max(self):
        return self.__current_value == self.__max_value
```

```python
def test_counter():
    my_counter = counter(max_value=3)
    print(f"Initial value: {my_counter.get_value()}")
    
    my_counter.increment()
    print(f"1 increment: {my_counter.get_value()}")

    my_counter.increment()
    print(f"2 increment: {my_counter.get_value()}")
    
    my_counter.increment()
    print(f"3 increment: {my_counter.get_value()}")
    
    my_counter.increment() 
    
    my_counter.reset()
    print(f"Reset Value: {my_counter.get_value()}")
    
    print("\nAttempting to access private data directly:")
    try:
        print(my_counter.__current_value)
    except AttributeError as e:
        print(f"Blocked access to private variable. Error caught: {e}")

if __name__ == "__main__":
    test_counter()
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
import math

class Circle:
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_area(self):
        return math.pi * (self.__radius ** 2)

    def get_perimeter(self):
        return 2 * math.pi * self.__radius
```

```python
def test_circle():
    
    my_circle = Circle(radius=5, x=0, y=0)
    area = my_circle.get_area()
    perimeter = my_circle.get_perimeter()
    
    print(f"Radius: {my_circle.get_radius()}")
    print(f"Center X: {my_circle.get_x()}")
    print(f"Center Y: {my_circle.get_y()}")
    print(f"Area: {area:.2f}")
    print(f"Perimeter: {perimeter:.2f}")
    
if __name__ == "__main__":
    test_circle()
```

5. Implement a common base class for the classes implemented in 3 and 4 above which implements all common methods as not implemented functions (virtual). Re-implement your regtangle and circule classes to inherit from the base class and overload the functions accordingly. 

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_area(self):
        raise NotImplementedError

    def get_perimeter(self):
        raise NotImplementedError

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)  
        self.__length = length
        self.__width = width

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_area(self):
        return self.__length * self.__width

    def get_perimeter(self):
        return 2 * (self.__length + self.__width)


class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y) 
        self.__radius = radius

    def get_radius(self):
        return self.__radius

    def get_area(self):
        return math.pi * (self.__radius ** 2)

    def get_perimeter(self):
        return 2 * math.pi * self.__radius
```

```python
def test_shapes():
    
    my_rect = Rectangle(length=5, width=10, x=2, y=3)
    my_circle = Circle(radius=5, x=0, y=0)
    
    print(f"Rectangle Origin: ({my_rect.get_x()}, {my_rect.get_y()})")
    print(f"Circle Origin: ({my_circle.get_x()}, {my_circle.get_y()})")
    
    shapes_list = [my_rect, my_circle]
    
    for shape in shapes_list:
        shape_name = shape.__class__.__name__ 
        print(f"{shape_name} Area: {shape.get_area():.2f}")
        print(f"{shape_name} Perimeter: {shape.get_perimeter():.2f}")

if __name__ == "__main__":
    test_shapes()
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
import math

class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        super().__init__(x, y)
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c

    def get_side_a(self):
        return self.__side_a

    def get_side_b(self):
        return self.__side_b

    def get_side_c(self):
        return self.__side_c

    def get_perimeter(self):
        return self.__side_a + self.__side_b + self.__side_c

    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))
```

```python
def test_shapes():
    my_triangle = Triangle(side_a=3, side_b=4, side_c=5, x=1, y=1) 

    print(f"Triangle Origin: ({my_triangle.get_x()}, {my_triangle.get_y()})")
    
    shapes_list = [my_triangle]
    
    for shape in shapes_list:
        shape_name = shape.__class__.__name__ 
        print(f"{shape_name} Area: {shape.get_area():.2f}")
        print(f"{shape_name} Perimeter: {shape.get_perimeter():.2f}")

if __name__ == "__main__":
    test_shapes()
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def get_area(self):
        raise NotImplementedError

    def get_perimeter(self):
        raise NotImplementedError

    def get_perimeter_points(self):
        raise NotImplementedError

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)  
        self.__length = length
        self.__width = width

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_area(self):
        return self.__length * self.__width

    def get_perimeter(self):
        return 2 * (self.__length + self.__width)

    def get_perimeter_points(self):
        points = []
        x, y = self.get_x(), self.get_y()
        l, w = self.get_length(), self.get_width()
        
        for i in range(4): points.append((x + i*(l/4), y))          
        for i in range(4): points.append((x + l, y + i*(w/4)))       
        for i in range(4): points.append((x + l - i*(l/4), y + w))   
        for i in range(4): points.append((x, y + w - i*(w/4)))       
        
        return points

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y) 
        self.__radius = radius

    def get_radius(self):
        return self.__radius

    def get_area(self):
        return math.pi * (self.__radius ** 2)

    def get_perimeter(self):
        return 2 * math.pi * self.__radius

    def get_perimeter_points(self):
        points = []
        x, y = self.get_x(), self.get_y()
        r = self.get_radius()
        
        for i in range(16):
            theta = i * (2 * math.pi / 16)
            px = x + r * math.cos(theta)
            py = y + r * math.sin(theta)
            points.append((px, py))
            
        return points

class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        super().__init__(x, y)
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c

    def get_side_a(self): return self.__side_a
    def get_side_b(self): return self.__side_b
    def get_side_c(self): return self.__side_c

    def get_perimeter(self):
        return self.__side_a + self.__side_b + self.__side_c

    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))

    def get_perimeter_points(self):
        points = []
        x, y = self.get_x(), self.get_y()
        a, b, c = self.get_side_a(), self.get_side_b(), self.get_side_c()
        
        v1 = (x, y)
        v2 = (x + a, y)
        
        cos_theta = (a**2 + c**2 - b**2) / (2 * a * c)
        sin_theta = math.sqrt(1 - cos_theta**2)
        v3 = (x + c * cos_theta, y + c * sin_theta)
        
        for i in range(6): 
            t = i / 6.0
            points.append((v1[0] + t*(v2[0]-v1[0]), v1[1] + t*(v2[1]-v1[1])))
            
        for i in range(5): 
            t = i / 5.0
            points.append((v2[0] + t*(v3[0]-v2[0]), v2[1] + t*(v3[1]-v2[1])))
            
        for i in range(5): 
            t = i / 5.0
            points.append((v3[0] + t*(v1[0]-v3[0]), v3[1] + t*(v1[1]-v3[1])))
            
        return points
```

```python
def test_all_shapes():
    my_rect = Rectangle(length=5, width=10, x=2, y=3)
    my_circle = Circle(radius=5, x=0, y=0)
    my_triangle = Triangle(side_a=3, side_b=4, side_c=5, x=1, y=1) 
    
    shapes_list = [my_rect, my_circle, my_triangle]
    
    for shape in shapes_list:
        shape_name = shape.__class__.__name__ 
        print(f"\n{shape_name}:")
        
        points = shape.get_perimeter_points()
        for count, (px, py) in enumerate(points, 1):
            print(f"    Pt {count:2}: ({px:>5.2f}, {py:>5.2f})")

if __name__ == "__main__":
    test_all_shapes()
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y

    def get_area(self): raise NotImplementedError
    def get_perimeter(self): raise NotImplementedError
    def get_perimeter_points(self): raise NotImplementedError

    def is_inside(self, px, py):
        """Virtual method to test if coordinates (px, py) are inside the shape."""
        raise NotImplementedError("Subclasses must implement is_inside().")


class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)  
        self.__length = length
        self.__width = width

    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_area(self): return self.__length * self.__width
    def get_perimeter(self): return 2 * (self.__length + self.__width)

    def get_perimeter_points(self):
        points = []
        x, y = self.get_x(), self.get_y()
        l, w = self.get_length(), self.get_width()
        
        for i in range(4): points.append((x + i*(l/4), y))           
        for i in range(4): points.append((x + l, y + i*(w/4)))       
        for i in range(4): points.append((x + l - i*(l/4), y + w))   
        for i in range(4): points.append((x, y + w - i*(w/4)))       
        return points

    def is_inside(self, px, py):
        """Checks if the point falls within the rectangle's x and y bounds."""
        x, y = self.get_x(), self.get_y()
        return (x <= px <= x + self.get_length()) and (y <= py <= y + self.get_width())


class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y) 
        self.__radius = radius

    def get_radius(self): return self.__radius
    def get_area(self): return math.pi * (self.__radius ** 2)
    def get_perimeter(self): return 2 * math.pi * self.__radius

    def get_perimeter_points(self):
        points = []
        x, y, r = self.get_x(), self.get_y(), self.get_radius()
        for i in range(16):
            theta = i * (2 * math.pi / 16)
            points.append((x + r * math.cos(theta), y + r * math.sin(theta)))
        return points

    def is_inside(self, px, py):
        """Checks if the distance from the point to the center is <= radius."""
        dx = px - self.get_x()
        dy = py - self.get_y()
        return (dx**2 + dy**2) <= (self.get_radius()**2)


class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        super().__init__(x, y)
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c

    def get_side_a(self): return self.__side_a
    def get_side_b(self): return self.__side_b
    def get_side_c(self): return self.__side_c
    def get_perimeter(self): return self.__side_a + self.__side_b + self.__side_c
    
    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))

    def _get_vertices(self):
        """Helper method to calculate the 3 vertices to avoid repetitive code."""
        x, y = self.get_x(), self.get_y()
        a, b, c = self.get_side_a(), self.get_side_b(), self.get_side_c()
        
        v1 = (x, y)
        v2 = (x + a, y)
        cos_theta = (a**2 + c**2 - b**2) / (2 * a * c)
        sin_theta = math.sqrt(1 - cos_theta**2)
        v3 = (x + c * cos_theta, y + c * sin_theta)
        return v1, v2, v3

    def get_perimeter_points(self):
        points = []
        v1, v2, v3 = self._get_vertices()
        
        for i in range(6): 
            t = i / 6.0
            points.append((v1[0] + t*(v2[0]-v1[0]), v1[1] + t*(v2[1]-v1[1])))
        for i in range(5): 
            t = i / 5.0
            points.append((v2[0] + t*(v3[0]-v2[0]), v2[1] + t*(v3[1]-v2[1])))
        for i in range(5): 
            t = i / 5.0
            points.append((v3[0] + t*(v1[0]-v3[0]), v3[1] + t*(v1[1]-v3[1])))
        return points

    def is_inside(self, px, py):
        """Uses the sub-triangle area method to check if a point is inside."""
        v1, v2, v3 = self._get_vertices()
        
        # Nested helper function to calculate the area from 3 raw coordinates
        def area_from_points(p1, p2, p3):
            return abs((p1[0]*(p2[1] - p3[1]) + p2[0]*(p3[1] - p1[1]) + p3[0]*(p1[1] - p2[1])) / 2.0)
            
        total_area = area_from_points(v1, v2, v3)
        sub1 = area_from_points((px, py), v1, v2)
        sub2 = area_from_points((px, py), v2, v3)
        sub3 = area_from_points((px, py), v3, v1)
        
        return math.isclose(total_area, sub1 + sub2 + sub3, rel_tol=1e-5)
```

```python
def test_is_inside():
    my_rect = Rectangle(length=5, width=10, x=2, y=3)
    print("Rectangle")
    print(f"Inside check (4, 5) - {my_rect.is_inside(4, 5)}")
    print(f"Outside check (10, 10) - {my_rect.is_inside(10, 10)}\n")

    my_circle = Circle(radius=5, x=0, y=0)
    print("Circle")
    print(f"Inside check (2, 2) - {my_circle.is_inside(2, 2)}")
    print(f"Outside check (5, 5) - {my_circle.is_inside(5, 5)}\n")

    my_triangle = Triangle(side_a=3, side_b=4, side_c=5, x=1, y=1) 
    print("Triangle")
    print(f"Inside check (2, 2) - {my_triangle.is_inside(2, 2)}")
    print(f"Outside check (4, 4) - {my_triangle.is_inside(4, 4)}\n")

if __name__ == "__main__":
    test_is_inside()
```

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x, x+w):
            self.set_pixel(i, y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y, y+h):
            self.set_pixel(x, i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        if x2 - x1 == 0:
            self.v_line(y1, x1, y2-y1, **kargs)
            return
        slope = (y2-y1) / (x2-x1)
        for y in range(y1, y2):
            x = int(slope * y)
            self.set_pixel(x, y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y

    def get_area(self): raise NotImplementedError
    def get_perimeter(self): raise NotImplementedError
    def get_perimeter_points(self): raise NotImplementedError
    def is_inside(self, px, py): raise NotImplementedError

    def paint(self, canvas, char='*'):
        for row in range(canvas.height):
            for col in range(canvas.width):
                if self.is_inside(col, row):
                    canvas.set_pixel(row, col, char)


class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)  
        self.__length = length
        self.__width = width

    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_area(self): return self.__length * self.__width
    def get_perimeter(self): return 2 * (self.__length + self.__width)

    def get_perimeter_points(self):
        points = []
        x, y, l, w = self.get_x(), self.get_y(), self.get_length(), self.get_width()
        for i in range(4): points.append((x + i*(l/4), y))           
        for i in range(4): points.append((x + l, y + i*(w/4)))       
        for i in range(4): points.append((x + l - i*(l/4), y + w))   
        for i in range(4): points.append((x, y + w - i*(w/4)))       
        return points

    def is_inside(self, px, py):
        x, y = self.get_x(), self.get_y()
        return (x <= px <= x + self.get_length()) and (y <= py <= y + self.get_width())

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y) 
        self.__radius = radius

    def get_radius(self): return self.__radius
    def get_area(self): return math.pi * (self.__radius ** 2)
    def get_perimeter(self): return 2 * math.pi * self.__radius

    def get_perimeter_points(self):
        points = []
        x, y, r = self.get_x(), self.get_y(), self.get_radius()
        for i in range(16):
            theta = i * (2 * math.pi / 16)
            points.append((x + r * math.cos(theta), y + r * math.sin(theta)))
        return points

    def is_inside(self, px, py):
        dx = px - self.get_x()
        dy = py - self.get_y()
        return (dx**2 + dy**2) <= (self.get_radius()**2)

class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        super().__init__(x, y)
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c

    def get_side_a(self): return self.__side_a
    def get_side_b(self): return self.__side_b
    def get_side_c(self): return self.__side_c
    def get_perimeter(self): return self.__side_a + self.__side_b + self.__side_c
    
    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))

    def get_vertices(self):
        x, y = self.get_x(), self.get_y()
        a, b, c = self.get_side_a(), self.get_side_b(), self.get_side_c()
        
        v1 = (x, y)
        v2 = (x + a, y)
        
        cos_theta = (a**2 + c**2 - b**2) / (2 * a * c)
        sin_theta = math.sqrt(1 - cos_theta**2)
        
        # FLIPPED: We subtract from 'y' so the tip points UP towards the top of the canvas
        v3 = (x + c * cos_theta, y - c * sin_theta)
        
        return v1, v2, v3

    def get_perimeter_points(self):
        points = []
        v1, v2, v3 = self._get_vertices()
        for i in range(6): 
            t = i / 6.0
            points.append((v1[0] + t*(v2[0]-v1[0]), v1[1] + t*(v2[1]-v1[1])))
        for i in range(5): 
            t = i / 5.0
            points.append((v2[0] + t*(v3[0]-v2[0]), v2[1] + t*(v3[1]-v2[1])))
        for i in range(5): 
            t = i / 5.0
            points.append((v3[0] + t*(v1[0]-v3[0]), v3[1] + t*(v1[1]-v3[1])))
        return points

    def is_inside(self, px, py):
        v1, v2, v3 = self.get_vertices()
        def area_from_points(p1, p2, p3):
            return abs((p1[0]*(p2[1] - p3[1]) + p2[0]*(p3[1] - p1[1]) + p3[0]*(p1[1] - p2[1])) / 2.0)
            
        total_area = area_from_points(v1, v2, v3)
        sub1 = area_from_points((px, py), v1, v2)
        sub2 = area_from_points((px, py), v2, v3)
        sub3 = area_from_points((px, py), v3, v1)
        
        return math.isclose(total_area, sub1 + sub2 + sub3, abs_tol=0.1)

class CompoundShape(Shape):
    def __init__(self):
        super().__init__(0, 0)
        self.__shapes = []

    def add_shape(self, shape):
        self.__shapes.append(shape)

    def is_inside(self, px, py):
        for shape in self.__shapes:
            if shape.is_inside(px, py):
                return True
        return False

    def paint(self, canvas, char='*'):
        for row in range(canvas.height):
            for col in range(canvas.width):
                if self.is_inside(col, row):
                    canvas.set_pixel(row, col, char)
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self):
        self.__shapes = []

    def add_shape(self, shape, char='*'):
        self.__shapes.append((shape, char))

    def remove_shape(self, shape):
        self.__shapes = [item for item in self.__shapes if item[0] != shape]

    def clear_drawing(self):
        self.__shapes = []

    def paint(self, canvas):
        for shape, char in self.__shapes:
            shape.paint(canvas, char)
```

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x, x+w):
            self.set_pixel(i, y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y, y+h):
            self.set_pixel(x, i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        if x2 - x1 == 0:
            self.v_line(y1, x1, y2-y1, **kargs)
            return
        slope = (y2-y1) / (x2-x1)
        for y in range(y1, y2):
            x = int(slope * y)
            self.set_pixel(x, y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y

    def get_x(self): return self.__x
    def get_y(self): return self.__y

    def get_area(self): raise NotImplementedError
    def get_perimeter(self): raise NotImplementedError
    def get_perimeter_points(self): raise NotImplementedError
    def is_inside(self, px, py): raise NotImplementedError

    def paint(self, canvas, char='*'):
        for row in range(canvas.height):
            for col in range(canvas.width):
                if self.is_inside(col, row):
                    canvas.set_pixel(row, col, char)

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)  
        self.__length = length
        self.__width = width

    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_area(self): return self.__length * self.__width
    def get_perimeter(self): return 2 * (self.__length + self.__width)

    def get_perimeter_points(self):
        points = []
        x, y, l, w = self.get_x(), self.get_y(), self.get_length(), self.get_width()
        for i in range(4): points.append((x + i*(l/4), y))           
        for i in range(4): points.append((x + l, y + i*(w/4)))       
        for i in range(4): points.append((x + l - i*(l/4), y + w))   
        for i in range(4): points.append((x, y + w - i*(w/4)))       
        return points

    def is_inside(self, px, py):
        x, y = self.get_x(), self.get_y()
        return (x <= px <= x + self.get_length()) and (y <= py <= y + self.get_width())

    def __repr__(self):
        return f"Rectangle({self.get_length()}, {self.get_width()}, {self.get_x()}, {self.get_y()})"

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y) 
        self.__radius = radius

    def get_radius(self): return self.__radius
    def get_area(self): return math.pi * (self.__radius ** 2)
    def get_perimeter(self): return 2 * math.pi * self.__radius

    def get_perimeter_points(self):
        points = []
        x, y, r = self.get_x(), self.get_y(), self.get_radius()
        for i in range(16):
            theta = i * (2 * math.pi / 16)
            points.append((x + r * math.cos(theta), y + r * math.sin(theta)))
        return points

    def is_inside(self, px, py):
        dx = px - self.get_x()
        dy = py - self.get_y()
        return (dx**2 + dy**2) <= (self.get_radius()**2)

    def __repr__(self):
        return f"Circle({self.get_radius()}, {self.get_x()}, {self.get_y()})"

class Triangle(Shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        super().__init__(x, y)
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c

    def get_side_a(self): return self.__side_a
    def get_side_b(self): return self.__side_b
    def get_side_c(self): return self.__side_c
    def get_perimeter(self): return self.__side_a + self.__side_b + self.__side_c
    
    def get_area(self):
        s = self.get_perimeter() / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))

    def _get_vertices(self):
        x, y = self.get_x(), self.get_y()
        a, b, c = self.get_side_a(), self.get_side_b(), self.get_side_c()
        v1 = (x, y)
        v2 = (x + a, y)
        cos_theta = (a**2 + c**2 - b**2) / (2 * a * c)
        sin_theta = math.sqrt(1 - cos_theta**2)
        v3 = (x + c * cos_theta, y - c * sin_theta) # Flipped for upward-pointing roof
        return v1, v2, v3

    def get_perimeter_points(self):
        points = []
        v1, v2, v3 = self._get_vertices()
        for i in range(6): 
            t = i / 6.0
            points.append((v1[0] + t*(v2[0]-v1[0]), v1[1] + t*(v2[1]-v1[1])))
        for i in range(5): 
            t = i / 5.0
            points.append((v2[0] + t*(v3[0]-v2[0]), v2[1] + t*(v3[1]-v2[1])))
        for i in range(5): 
            t = i / 5.0
            points.append((v3[0] + t*(v1[0]-v3[0]), v3[1] + t*(v1[1]-v3[1])))
        return points

    def is_inside(self, px, py):
        v1, v2, v3 = self._get_vertices()
        def area_from_points(p1, p2, p3):
            return abs((p1[0]*(p2[1] - p3[1]) + p2[0]*(p3[1] - p1[1]) + p3[0]*(p1[1] - p2[1])) / 2.0)
            
        total_area = area_from_points(v1, v2, v3)
        sub1 = area_from_points((px, py), v1, v2)
        sub2 = area_from_points((px, py), v2, v3)
        sub3 = area_from_points((px, py), v3, v1)
        return math.isclose(total_area, sub1 + sub2 + sub3, abs_tol=0.1)

    def __repr__(self):
        return f"Triangle({self.get_side_a()}, {self.get_side_b()}, {self.get_side_c()}, {self.get_x()}, {self.get_y()})"

class CompoundShape(Shape):
    def __init__(self, shapes=None):
        super().__init__(0, 0)
        self.__shapes = shapes if shapes is not None else []

    def add_shape(self, shape):
        self.__shapes.append(shape)

    def is_inside(self, px, py):
        for shape in self.__shapes:
            if shape.is_inside(px, py):
                return True
        return False

    def paint(self, canvas, char='*'):
        for row in range(canvas.height):
            for col in range(canvas.width):
                if self.is_inside(col, row):
                    canvas.set_pixel(row, col, char)

    def __repr__(self):
        return f"CompoundShape({repr(self.__shapes)})"

class RasterDrawing:
    def __init__(self, shapes=None):
        self.__shapes = shapes if shapes is not None else []

    def add_shape(self, shape, char='*'):
        self.__shapes.append((shape, char))

    def remove_shape(self, shape):
        self.__shapes = [item for item in self.__shapes if item[0] != shape]

    def clear_drawing(self):
        self.__shapes = []

    def paint(self, canvas):
        for shape, char in self.__shapes:
            shape.paint(canvas, char)

    def __repr__(self):
        return f"RasterDrawing({repr(self.__shapes)})"

    def save(self, filename):
        with open(filename, "w") as f:
            f.write(repr(self))

def load_raster_drawing(filename):
    with open(filename, "r") as f:
        reconstructed_drawing = eval(f.read())
    return reconstructed_drawing
```
